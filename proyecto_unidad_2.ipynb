c) Modelo ANOVA
Aplica un ANOVA (Análisis de Varianza) considerando todas las variables disponibles.

Interpreta los resultados y comenta cuáles variables resultan significativas para explicar la variable objetivo.

Incluye una tabla con los valores F y los p-valores.



```python
#con este codigo vamos a ver f y el valor p de todas las variables, con el valor p vamos a saber si es hay diferencia significativa entre las variables respecto al precio
#ho- no hay diferencia significativa
#h1- si hay diferencia significativa
#nivel de significancia: 0.05
#si el valor p es menosr a 0.05 ho se rechaza por lo que hay diferencia significativa y se puede utilizar la variable en el modelo
#se analiza variable por variable el valor p y se rechaza o acepta h0
import statsmodels.api as sm
from statsmodels.formula.api import ols
# Y ~ X
modelo_lineal = ols('precio ~ costo + ventas + margen_beneficio + inventario + porcentaje_descuento + dia_entrega + C(categoria) + C(material) + C(color) + C(ubicacion) + C(temporada) + C(tipo_tienda) + C(marca) + ingresos', data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

d) Test de Tukey y codificación
Para las variables categóricas que hayan resultado significativas en el ANOVA:

Aplica la prueba de Tukey HSD.

Realiza el encoding (codificación) respetando el orden que dicta Tukey, según los grupos estadísticamente diferentes.



```python
#ya con anova hecho ahora damos paso a hacer tukie, en  este caso realizamos todas, sin embargo con el paso anterior pudimos descartar los valores que no necesitamos o que no nos dan diferencia significativa
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(
    endog=df['precio'], # Variable de respuesta
    groups=df['categoria'],          # Variable categorica
    alpha=nivel_de_significancia) # nivel de significancia
print(tukey)

tukey.plot_simultaneous(ylabel="categoria", xlabel="precio")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

f) Modelo matemático de precio de venta
Obtén un modelo matemático de regresión múltiple que permita estimar el precio de venta del mueble.

Presenta la ecuación con los coeficientes y explica brevemente el significado de cada término.

```python
#tomamos las variables que nos aportan información, en este caso eliminamos categorias segun nos indican tres factores, el primero el resultado del valor p según ANOVA, en segundo lugar si hay multicolinealidad y como tercer factor para eliminar el no conocimiento del factor
x = df[["categoria","material","color","ubicacion","temporada", "marca", "margen_beneficio", "ingresos", "costo"]]
y = df["precio"]


df_cleaned = df[x.columns.tolist() + [y.name]].dropna()
x_cleaned = df_cleaned[x.columns]
y_cleaned = df_cleaned[y.name]

#Recta de regresión ajustada
import statsmodels.api as sm
x_constante = sm.add_constant(x_cleaned)
modelo = sm.OLS(y_cleaned, x_constante).fit()

yc = modelo.predict(x_constante)

#Coeficiente de determinación
from sklearn.metrics import r2_score  # recomendada
r2 = r2_score(y_cleaned, yc)
print(f'Coeficiente de determinación: {r2: 0.4f}\n')

# 12. Realice una tabla ANOVA e interprete el resultado.
from statsmodels.formula.api import ols
# Y ~ X
modelo_lineal = ols('precio ~ categoria + material + color + ubicacion + temporada + marca + margen_beneficio + ingresos + costo', data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

g) Evaluación del modelo Analiza los residuales del modelo y los estadísticos principales, especialmente el coeficiente de determinación (R²). Comenta si el modelo cumple con los supuestos de linealidad, homocedasticidad y normalidad de errores.

```python
x = df[['categoria']]
y = df["precio"]


df_cleaned = df[x.columns.tolist() + [y.name]].dropna()
x_cleaned = df_cleaned[x.columns]
y_cleaned = df_cleaned[y.name]

#Recta de regresión ajustada
import statsmodels.api as sm
x_constante = sm.add_constant(x_cleaned)
modelo = sm.OLS(y_cleaned, x_constante).fit()

yc = modelo.predict(x_constante)

#Coeficiente de determinación
from sklearn.metrics import r2_score  # recomendada
r2 = r2_score(y_cleaned, yc)
print(f'Coeficiente de determinación: {r2: 0.4f}\n')

residuales = modelo.resid
plt.figure()
plt.scatter(x_cleaned['categoria'], residuales, color = 'blue')
plt.xlabel('categoria', fontsize = 12)
plt.ylabel('Residuales', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales vs categoria", fontsize = 16)
```
